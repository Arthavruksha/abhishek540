# Q1. Create 500 txt files in a directory. Every file contains 20,000 lines and every line contain random string of length 20 characters.

```python
import os
path = os.getcwd()
print(path)
```

```python
import string
import secrets
import random
import time

def random_20_char():
    X = ""
    N = random.randint(20,21)
    res = "".join(secrets.choice(string.ascii_letters) for i in range(N))
#     print(res)
    return res
```

```python
time_dict = {"File_number":[],"time_create":[],"time_upper":[]}
```

```python
for i in range(500):
    x = time.time()
    file_path = path+"/"+"file_"+str(i)+".txt"
    print(file_path)
    print("Creating...")
    f= open(file_path,"w+")
    print("Created!!!")
    for i in range(20000):
        file_string = random_20_char()
        f.write(file_string+"\n")
    y = time.time()
    time_dict['File_number'].append(i)
    time_dict['time_create'].append(y-x)
        
```

# Q2. Calculate the execution time to convert all the file to upper case. Save the results in csv file as given below.

```python


for i in range(500):
    begin = time.time()
    file_path = path+"/"+"file_"+str(i)+".txt"
    print(file_path)
    print("Reading File ----> "+ file_path)
    with open(file_path,"r+b") as input_file:
        content = input_file.read()
        input_file.seek(0)
        input_file.write(content.upper())
        input_file.close()
        
    end = time.time()
    time_consumed = end-begin
    print(i)
    time_dict['time_upper'].append(time_consumed)
    print("Time to convert the content of file in UPPERCASE --->",time_consumed)
print(time_dict)

#time.sleep()

    
```

```python
import pandas as pd

df = pd.DataFrame.from_dict(time_dict)
  
print(df)
    
df.to_csv("time_consumed.csv")
```
