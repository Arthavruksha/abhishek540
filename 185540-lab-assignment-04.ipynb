```python
df_length = len(df)

column_result = ["SN","F1","F2","F3","F4","F5","F6","Class"]
column_log = ["File_name","Sequence","Class"]

result = pd.DataFrame(columns = column_result)
log = pd.DataFrame(columns = column_log)

for i in range(df_length):
#     print(df.iloc[i]["Class"])
#     print(df.iloc[i]["Sequence"])
    class_value = df.iloc[i]["Class"]
    seq = df.iloc[i]["Sequence"]
    if pd.isna(class_value)  or pd.isna(seq) or any(map(str.isdigit,seq)) :
        log_x = {"File_name":file_name_final,"Sequence":seq,"Class":class_value}
        print(log_x)
        log = log.append(log_x , ignore_index= True)

    else:
        SN = i
        F1 = seq.count("N")
        F2 = seq.count("H")
        F3 = seq.count("Q")
        F4 = seq.count("G")
        F5 = seq.count("D")
        F6 = seq.count("T")
        Class = 1 if class_value=="+" else 0
        result_x = {"SN":SN,"F1":F1,"F2":F2,"F3":F3,"F4":F4,"F5":F5,"F6":F6,"Class":Class}
        result = result.append(result_x,ignore_index = True)
    
    
```

```python
s = str(datetime.datetime.now())
s = s.replace(":","")
s = s.replace("-","")
s = s.replace(".","")
s = s.replace(" ","")
```

```python
output_file_name = s+".csv"
result.to_csv("result_"+output_file_name)
log.to_csv("log_"+output_file_name)
```
